## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict
from transformers import AutoTokenizer

class BPETokenizer:
    """
    Byte Pair Encoding Tokenizer class.

    Args:
        vocab_size (int): Size of the vocabulary.
    """
    def __init__(self, vocab_size):
        # Initialize the tokenizer with the specified vocabulary size.
        self.vocab_size = vocab_size
        self.tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
        self.word_freqs = defaultdict(int)
        self.splits = {}
        self.merges = {}

    def train(self, corpus):
        """
        Train the Byte Pair Encoding tokenizer on a given corpus.

        Args:
            corpus (List[str]): List of texts for training.

        Returns:
            dict: Merges dictionary.
        """
        self.corpus = corpus

        # Count the frequencies of individual words in the corpus.
        for text in self.corpus:
            words_with_offsets = self.tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(text)
            new_words = [word for word, offset in words_with_offsets]
            for word in new_words:
                self.word_freqs[word] += 1

        alphabet = []
        # Build the alphabet from the unique letters in the words.
        for word in self.word_freqs.keys():
            for letter in word:
                if letter not in alphabet:
                    alphabet.append(letter)
        alphabet.sort()

        vocab = ["</w>"] + alphabet.copy()
        self.splits = {word: [c for c in word] for word in self.word_freqs.keys()}

        # Continue merging pairs until the vocabulary size is reached.
        while len(vocab) < self.vocab_size:
            pair_freqs = self.compute_pair_freqs()

            # Find the most frequent pair in the computed pair frequencies.
            best_pair = ""
            max_freq = None
            for pair, freq in pair_freqs.items():
                if max_freq is None or max_freq < freq:
                    best_pair = pair
                    max_freq = freq

            # Merge the best pair and update the splits and merges dictionaries.
            self.splits = self.merge_pair(*best_pair)
            self.merges[best_pair] = best_pair[0] + best_pair[1]
            vocab.append(best_pair[0] + best_pair[1])
        return self.merges

    def compute_pair_freqs(self):
        """
        Compute pair frequencies in the current splits.

        Returns:
            defaultdict: Pair frequencies dictionary.
        """
        pair_freqs = defaultdict(int)
        for word, freq in self.word_freqs.items():
            split = self.splits[word]
            if len(split) == 1:
                continue
            # Count the frequencies of pairs in each word.
            for i in range(len(split) - 1):
                pair = (split[i], split[i + 1])
                pair_freqs[pair] += freq
        return pair_freqs

    def merge_pair(self, a, b):
        """
        Merge a pair in the splits.

        Args:
            a (str): First part of the pair.
            b (str): Second part of the pair.

        Returns:
            dict: Updated splits dictionary.
        """
        for word in self.word_freqs:
            split = self.splits[word]
            if len(split) == 1:
                continue
            i = 0
            # Merge the specified pair in each word's split.
            while i < len(split) - 1:
                if split[i] == a and split[i + 1] == b:
                    split = split[:i] + [a + b] + split[i + 2 :]
                else:
                    i += 1
            self.splits[word] = split
        return self.splits

    def tokenize(self, text):
        """
        Tokenize the input text using the trained tokenizer.

        Args:
            text (str): Input text.

        Returns:
            List[str]: List of tokens.
        """
        pre_tokenize_result = self.tokenizer._tokenizer.pre_tokenizer.pre_tokenize_str(text)
        pre_tokenized_text = [word for word, offset in pre_tokenize_result]
        splits_text = [[l for l in word] for word in pre_tokenized_text]

        # Apply the merges to tokenize the input text.
        for pair, merge in self.merges.items():
            for idx, split in enumerate(splits_text):
                i = 0
                while i < len(split) - 1:
                    if split[i] == pair[0] and split[i + 1] == pair[1]:
                        split = split[:i] + [merge] + split[i + 2 :]
                    else:
                        i += 1
                splits_text[idx] = split
        result = sum(splits_text, [])
        return result
```

```python
import os
import matplotlib.pyplot as plt
```

```python
def load_tweets(file_path):
    """
    Load tweets from a file.

    Args:
    file_path (str): Path to the file containing tweets.

    Returns:
    List[str]: List of tweets.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        tweets = file.readlines()
    return tweets
```

```python
def analyze_vocab_size_effect(vocab_sizes, tweet_files):
    """
    Analyze the effect of vocab_size on the average length of tokenized tweets.

    Args:
    vocab_sizes (List[int]): List of vocab_size values to analyze.
    tweet_files (List[str]): List of paths to tweet files.

    Returns:
    None (Plots the results).
    """
    avg_lengths = []

    for vocab_size in vocab_sizes:
        tokenizer = BPETokenizer(vocab_size=vocab_size)
        corpus = []

        # Load and concatenate tweets from all files
        for tweet_file in tweet_files:
            file_path = os.path.join(os.path.curdir, 'data', tweet_file)
            corpus.extend(load_tweets(file_path))

        # Train the tokenizer and tokenize tweets
        tokenizer.train(corpus)
        tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in corpus]

        # Calculate average length of tokenized tweets
        avg_length = sum(len(tokens) for tokens in tokenized_tweets) / len(tokenized_tweets)
        avg_lengths.append(avg_length)

    # Plot the results
    plt.plot(vocab_sizes, avg_lengths, marker='o')
    plt.title('Effect of Vocab Size on Average Tokenized Tweet/Sentence Length')
    plt.xlabel('Vocab Size')
    plt.ylabel('Average Tokenized Tweet/Sentence Length')
    plt.show()
```

```python
base_vocab_size = 100
vocab_sizes_to_analyze = [base_vocab_size, 250, 500, 750, 1000]
tweet_files_to_analyze = ['file1', 'file2', 'file3', 'file4', 'file5']

analyze_vocab_size_effect(vocab_sizes_to_analyze, tweet_files_to_analyze)
```
